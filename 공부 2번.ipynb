```python
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' : # row[-1] 값이 공백이 아닐 때
        if row[0].split('-')[1] == '08' : # row[0] 값을 '-'으로 분리하여 1번째 인덱스 값이 '08'이면
            result.append(float(row[-1])) # row[-1] 값을 실수로 변환하여 result에 추가한다
            
import matplotlib.pyplot as plt
plt.plot(result, 'hotpink')
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if 1998 <= int(row[0].split('-')[0]) : # 비교 연산자 대소 비교를 하기 위해 숫자 데이터로 형변환 해줌
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '28' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
import matplotlib.pyplot as plt
plt.plot(high, 'red')
plt.plot(low, 'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '' :
        date = row[0].split('-') # 날짜 값을 '-' 문자로 구분해서 저장
        if 1998 <= int(date[0]) : # 1998년 이후 데이터라면
            if date[1] == '02' and date[2] == '28' : # 2월 28일이라면
                high.append(float(row[-1]))
                low.append(float(row[-2]))
                
plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정 (제목 글자 깨짐 방지)
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프') 
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.hist(aug, bins = 100, color = 'r')
plt.show()
# 1907~2018년까지 8월달의 최고기온을 분석해본 결과 약 28~32도가 가장 많았다
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```
