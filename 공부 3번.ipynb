```python
import csv

f = open('age.csv')
data = csv.reader(f)
result = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(int(i)) # 정수형으로 형변환을 하고 result에 값 추가한다.
print(result)

# -- 신도림 데이터를 검색 -> 검색 결과에서 3~마지막열 즉, 0세~100세까지의 각 인구수를 가지고 와서 
# result에 각 인구수를 정수형으로 형변환하여 저장
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot') # 그래프 배경에 격자 무늬를 넣어주는 스타일 적용
plt.plot(result)
plt.show()
```

```python
import csv

f = open('age.csv')
data = csv.reader(f)
result = []
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요: ")

for row in data :
    if name in row[0] :
        for i in row[3:] :
            result.append(int(i))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data :
    if '신도림' in row[0] :
        for i in range(0,101):
            m.append(int(row[i+3])) # 남자 0세~100세 : 3번 열부터 103번 열까지의 데이터를 구하기 위해 i+3 
            f.append(int(row[-(i+1)])) # 여자 0세~100세 : -1번 열부터 -101번 열까지의 데이터를 구하기 위해 -(i+1)
f.reverse() # 여자 데이터가 100세부터 0세까지의 순서로 저장되어 있어서 값을 뒤집어줌
```

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] : 
            m.append(-int(i)) # 항아리 모양으로 그래프를 그리기 위해 남자 데이터를 음수로 바꿔줌
        for i in row[106:] :
            f.append(int(i))
```

```python
import matplotlib.pyplot as plt

plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data : 
    if name in row[0] :
        for i in row[3:104] :
            m.append(int(i))
        for i in row[106:] :
            f.append(int(i))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
