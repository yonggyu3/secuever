```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
# 제주특별자치도 검색 -> gender.csv 파일에 제주특별자치도라는 데이터가 여러 개 존재해서 오류

for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i))
        for i in row[106:] :
            f.append(int(i))
        break # 가장 처음에 나오는 제주특별자치도 데이터를 찾고 반복을 멈춤

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()            
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) : # 0~100까지 반복
            m += int(row[i+3]) # 3~103번열의 데이터. 남성 0세~100세
            f += int(row[i+106]) # 106~206번열의 데이터. 여성 0세~100세
            # 리스트가 아닌 변수에 값을 누적해서 저장
            # 0세부터 100세까지의 남성의 총 인구, 여성의 총 인구를 알 수 있음
        break
size.append(m)
size.append(f) # size[0] 남성의 총 인구, size[1] 여성의 총 인구

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['남','여'], autopct = '%.1f%%', colors = color, startangle = 90) 
# startangle = 90 기존 3시 기준에서 반시계 방향으로 90도 회전해서 기준점을 잡음
plt.title(name + ' 지역의 남녀 성별 비율')
plt.legend()
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('찾고 싶은 동네를 입력하세요 : ')

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            result.append(int(row[i]) - int(row[i+103])) # 각 연령대별 (남성 인구수 - 여성 인구수) 계산해서 result에 저장 
        break
        
import matplotlib.pyplot as plt
plt.bar(range(101), result) # result 값이 양수면 남자 인구수가 많고, 값이 음수면 여성 인구수가 많다
plt.show()
```

```python
import csv
import math
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []
name = input('굼긍한 동네를 입력하세요 : ')

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i]) + int(row[i+103]))) # math.sqrt 제곱근. 총인구수의 제곱근을 구함
        break
    
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (10,5), dpi = 300)
plt.title(name + ' 지역의 성별 인구 그래프')
plt.scatter(m, f, s = size, c = range(101), cmap = 'CMRmap', alpha = 0.5) # 총 인구수가 많을수록 점 크기가 커짐
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
