```python
# Q. 강릉시 지역에서 영유아(0~10세) 인구수가 제일 많은 동네는 어디일까?

import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)
next(data)

city = [] # 강릉시 읍/면/동 저장할 리스트
j = [] # 강릉시 행정구역 별 0세~10세 데이터를 저장할 리스트
a = a1 = a2 = a3 = a4 = a5 = a6 = a7 = a8 = a9 = a10 = a11 = 0
a12 = a13 = a14 = a15 = a16 = a17 = a18 = a19 = a20 = 0 # 행정구역 별 0~10세 인구수를 저장할 변수
young = [] # 행정구역 별 0세~10세 인구수를 저장할 리스트


for row in data:
    name = row[0].split('(') # name에 행정구역을 '(' 기준으로 구분해서 저장
    name2 = name[0].split() # name2[0]을 공백 기준으로 구분해서 저장

    if len(name2) == 3: # name2의 길이가 3개이면
        if name2[1] == '강릉시': # name2[1]이 강릉시이면
            city.append(name2[2]) # city에 강릉시 읍/면/동만 추출된 행정구역 저장
            j.append(row[3:14]) # j에 강릉시 행정구역 별 0세~10세 데이터 추가
                
for i in range(11): # row[3:14] 0세부터 10세까지의 데이터를 더한 값을 행정구역 별로 각 변수에 저장
    a += int(j[0][i]) # 강릉시 첫번째 행인 주문진읍의 0세~10세까지의 데이터를 더한 값 
    a1 += int(j[1][i]) # 강릉시 두번째 행인 성산면의 0세~10세까지의 데이터를 더한 값
    a2 += int(j[2][i]) # 반복문을 쓰고 싶었는데 실패
    a3 += int(j[3][i])
    a4 += int(j[4][i])
    a5 += int(j[5][i])
    a6 += int(j[6][i])
    a7 += int(j[7][i])
    a8 += int(j[8][i])
    a9 += int(j[9][i])
    a10 += int(j[10][i])
    a11 += int(j[11][i])
    a12 += int(j[12][i])
    a13 += int(j[13][i])
    a14 += int(j[14][i])
    a15 += int(j[15][i])
    a16 += int(j[16][i])
    a17 += int(j[17][i])
    a18 += int(j[18][i])
    a19 += int(j[19][i])
    a20 += int(j[20][i])
    
        
young.append(a) # young에 각 행정구역 별 0~10세 인구수를 추가
young.append(a1)
young.append(a2)
young.append(a3)
young.append(a4)
young.append(a5)
young.append(a6)
young.append(a7)
young.append(a8)
young.append(a9)
young.append(a10)
young.append(a11)
young.append(a12)
young.append(a13)
young.append(a14)
young.append(a15)
young.append(a16)
young.append(a17)
young.append(a18)
young.append(a19)
young.append(a20)


plt.figure(dpi=500) # 해상도 설정
plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정 (제목 글자 깨짐 방지)
plt.title('강릉시 지역의 영유아(0~10세) 인구 분포도') # 제목 설정
plt.style.use('seaborn-white') # 배경 스타일 설정
plt.bar(range(len(city)), young, color = 'steelblue') # 막대 그래프
plt.xticks(range(len(city)), labels = city) # 눈금 표시
plt.tick_params(axis = 'x', labelsize = 4) # 눈금 스타일 적용, axis x축 지정, labelsize 글자크기 4
plt.show()
print('강릉시 지역에서 영유아 인구수가 제일 많은 동네는', city[19]+'이고,', city[19]+'의 인구수는', young[19], '명으로 분석되었다.')

# Q. 강릉시 지역에서 영유아(0~10세) 인구수가 제일 많은 동네는 어디일까?
# 강릉시 지역에서 영유아 인구수가 제일 많은 동네는 성덕동이고, 성덕동의 인구수는 3133 명으로 분석되었다.
```
