```python
# Q. 유임승차 비율이 가장 높은 역은 어디이고, 그 역의 유/무임 승차 비율, 유/무임 하차 비율은 얼마나 차이날까?

import csv
import matplotlib.pyplot as plt
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

rate = 0 # 유임승차 비율
mx = 0 # 최대 유임승차 비율

rate3 = 0 # 무임승차 비율
rate4 = 0 # 유임하차 비율
rate5 = 0 # 무임하차 비율
rate_y = 0 # 유임승차 비율 - 무임승차 비율
rate_m = 0 # 무임하차 비율 - 무임하차 비율

label = ['유임승차','무임승차',] # label 이름
label2 = ['유임하차', '무임하차']
c = ['#14CCC0', '#B0E0E6'] # 색상 헥스 코드
c2 = ['#FF5C6A', '#FFB6C1']
plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정 (제목 글자 깨짐 방지)

for row in data:
    for i in range(4,8):
        row[i] = int(row[i]) # 4번열부터 7번열까지 정수형으로 형변환
        
    if row[6] != 0 and (row[4] + row[6]) > 1000000: # 무임승차 인원수가 0이 아니고, (유임승차 인원 + 무임승차 인원)이 백만명이 넘을 때 
        rate = row[4] / (row[4] + row[6]) # rate에 유임승차 비율 저장, 유임승차 인원 / (유임승차 인원 + 무임승차 인원)
        
        if rate > mx: # 최근 찾은 rate 값이 mx에 들어있는 값보다 크면
            mx = rate # mx에 rate 값 저장
            mx_station = row[3] + ' ' + row[1] # 최대 rate 값을 가지는 지하철역과 호선명을 저장
 
        if mx_station.split()[0] in row[3] and mx_station.split()[1] in row[1]: 
        # 최대 유임승차 비율을 가지는 지하철역과 호선명이 저장된 mx_station, 일치하는 데이터를 찾아서
            rate3 = row[6] / (row[4] + row[6]) # rate3에 무임승차 비율 저장
            rate4 = row[5] / (row[5] + row[7]) # rate4에 유임하차 비율 저장
            rate5 = row[7] / (row[5] + row[7]) # rate5에 무임하차 비율 저장
            rate_y = mx - rate3 # 유임승차 비율 - 무임승차 비율
            rate_m = rate4 - rate5 # 무임하차 비율 - 무임하차 비율
                                                            
        plt.title(row[3] + ' ' + row[1]) # 제목 설정
        plt.axis('equal') # 축 범위 지정
        plt.pie(row[4:8:2], labels = label, colors = c, autopct = '%.2f%%', startangle = 90) 
        # 원 그래프, autopct 소수점 2자리까지 나오게 설정, 시작각도 12시 방향부터 설정
        plt.legend()
        plt.savefig(row[3] + ' ' + row[1] + ' 승차비율.png') # savefig를 통해서 png파일로 저장
        plt.show() 
        plt.axis('equal')
        plt.pie(row[5:8:2], labels = label2, colors = c2, autopct = '%.2f%%', startangle = 90)
        plt.legend()
        plt.savefig(row[3] + ' ' + row[1] + ' 하차비율.png')
        plt.show()
        

print('유임승차 비율이 가장 높은 역은',mx_station+'으로, 유임승차 비율이', round(mx*100,2), '%이다.')
print(mx_station+'의 유임승차 비율은', round(mx*100,2), '%, 무임승차 비율은', round(rate3*100,2), '%로 유임승차 비율이', round(rate_y*100,2), '% 높게 나타났고,')
print('유임하차 비율은', round(rate4*100,2), '%, 무임하차 비율은', round(rate5*100,2), '%로 유임하차 비율이', round(rate_m*100,2), '%만큼 높게 나타났다.')

# Q. 유임승차 비율이 가장 높은 역은 어디이고, 그 역의 유/무임 승차 비율, 유/무임 하차 비율은 얼마나 차이날까?
  
# 유임승차 비율이 가장 높은 역은 홍대입구 2호선으로, 유임승차 비율이 95.34 %이다.
# 홍대입구 2호선의 유임승차 비율은 95.34 %, 무임승차 비율은 4.66 %로 유임승차 비율이 90.69 % 높게 나타났고,
# 유임하차 비율은 95.74 %, 무임하차 비율은 4.26 %로 유임하차 비율이 91.49 %만큼 높게 나타났다.
```
