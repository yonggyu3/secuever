```python
# Q. (7~9)      ,     ?

import csv
f = open('subwaytime.csv')
data = csv.reader(f)
next(data) 
next(data) 

result = [] #    
result2 = [] #    

mx = 0 #   
mx2 = 0 #   

mx_station = '' #     
mx_station2 = '' #      

for row in data:
    row[4:] = map(int, row[4:]) # map()  row[4:]  4   
    result.append(sum(row[10:15:2])) # row[10], row[12], row[14]  7 , 8 , 9     result 
    result2.append(sum(row[9:14:2])) # row[9], row[11], row[13]  7 , 8 , 9     result 
      
    if sum(row[10:15:2]) > mx: #   sum(row[10:15:2])  mx  
        mx = sum(row[10:15:2]) # mx sum(row[10:15:2])  .    
        mx_station = row[3] + '(' + row[1] + ')' #  mx     
    if sum(row[9:14:2]) > mx2: #   sum(row[9:14:2])  mx2 
        mx2 = sum(row[9:14:2]) # mx2 sum(row[9:14:2])  .    
        mx_station2 = row[3] + '(' + row[1] + ')' #  mx2     

result.sort() # result   
result2.sort() # result2   
        
import matplotlib.pyplot as plt  
plt.rc('font', family = 'Malgun Gothic') #      (   )
plt.style.use('ggplot') #     
plt.figure(dpi = 300) #  
plt.title('(7~9)     ') #  
plt.bar(range(len(result)), result) #  , x y 
plt.xlabel('', size = 'small', loc = 'right') # x  , size  , loc 
plt.ylabel('', size = 'small', rotation = 'horizontal' ,loc = 'top') # y  , rotation  
plt.show()
print('     ', mx_station,'', mx,'.')
plt.figure(dpi = 300) #  
plt.title('(7~9)     ') #  
plt.bar(range(len(result2)), result2) #  , x y 
plt.xlabel('', size = 'small', loc = 'right') # x  , size  , loc 
plt.ylabel('', size = 'small', rotation = 'horizontal' ,loc = 'top') # y  , rotation  
plt.show()
print('     ', mx_station2,'', mx2, '.')

# Q. (7~9)      ,     ?

#       (2)  809541 .
#       (2)  621974 .
```
