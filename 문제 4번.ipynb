```python
# Q. 출근시간(7시~9시)에 사람들이 가장 많이 승차하는 역은 어디이고, 가장 많이 하차하는 역은 어디일까?

import csv
f = open('subwaytime.csv')
data = csv.reader(f)
next(data) 
next(data) 

result = [] # 승차 인원을 저장해줄 리스트
result2 = [] # 하차 인원을 저장해줄 리스트

mx = 0 # 최대 승차 인원
mx2 = 0 # 최대 하차 인원

mx_station = '' # 최대 승차 인원을 가지는 지하철역
mx_station2 = '' # 최대 하차 인원을 가지는 지하철역 

for row in data:
    row[4:] = map(int, row[4:]) # map()함수를 사용해서 row[4:]이 가리키는 4번열부터 마지막열까지 정수형으로 형변환
    result.append(sum(row[10:15:2])) # row[10], row[12], row[14]가 가리키는 7시 승차, 8시 승차, 9시 승차 인원을 더한 값을 result에 추가
    result2.append(sum(row[9:14:2])) # row[9], row[11], row[13]가 가리키는 7시 하차, 8시 하차, 9시 하차 인원을 더한 값을 result에 추가
      
    if sum(row[10:15:2]) > mx: # 최근 찾은 sum(row[10:15:2]) 값이 mx보다 크다면 
        mx = sum(row[10:15:2]) # mx에 sum(row[10:15:2]) 값을 저장. 가장 큰 값이 저장됨
        mx_station = row[3] + '(' + row[1] + ')' # 최대 mx 값을 가지는 지하철역과 호선명 저장
    if sum(row[9:14:2]) > mx2: # 최근 찾은 sum(row[9:14:2]) 값이 mx2보다 크다면
        mx2 = sum(row[9:14:2]) # mx2에 sum(row[9:14:2]) 값을 저장. 가장 큰 값이 저장됨
        mx_station2 = row[3] + '(' + row[1] + ')' # 최대 mx2 값을 가지는 지하철역과 호선명 저장

result.sort() # result 값 오름차순 정렬
result2.sort() # result2 값 오름차순 정렬
        
import matplotlib.pyplot as plt  
plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정 (제목 글자 깨짐 방지)
plt.style.use('ggplot') # 배경에 격자 무늬 스타일 설정
plt.figure(dpi = 300) # 해상도 설정
plt.title('출근시간(7시~9시) 기준 역 별 승차 인원수') # 제목 설정
plt.bar(range(len(result)), result) # 막대 그래프, x축 y축 설정
plt.xlabel('지하철', size = 'small', loc = 'right') # x축 레이블 설정, size 텍스트 크기, loc 위치
plt.ylabel('인원수', size = 'small', rotation = 'horizontal' ,loc = 'top') # y축 레이블 설정, rotation 텍스트 회전각
plt.show()
print('출근시간에 사람들이 가장 많이 승차하는 역은', mx_station,'역으로', mx,'명이다.')
plt.figure(dpi = 300) # 해상도 설정
plt.title('출근시간(7시~9시) 기준 역 별 하차 인원수') # 제목 설정
plt.bar(range(len(result2)), result2) # 막대 그래프, x축 y축 설정
plt.xlabel('지하철', size = 'small', loc = 'right') # x축 레이블 설정, size 텍스트 크기, loc 위치
plt.ylabel('인원수', size = 'small', rotation = 'horizontal' ,loc = 'top') # y축 레이블 설정, rotation 텍스트 회전각
plt.show()
print('출근시간에 사람들이 가장 많이 하차하는 역은', mx_station2,'역으로', mx2, '명이다.')

# Q. 출근시간(7시~9시)에 사람들이 가장 많이 승차하는 역은 어디이고, 가장 많이 하차하는 역은 어디일까?

# 출근시간에 사람들이 가장 많이 승차하는 역은 신림(2호선) 역으로 809541 명이다.
# 출근시간에 사람들이 가장 많이 하차하는 역은 강남(2호선) 역으로 621974 명이다.
```
